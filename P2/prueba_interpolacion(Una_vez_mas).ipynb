```python
%pylab inline
from scipy.interpolate import griddata
from netCDF4 import Dataset
from mpl_toolkits.basemap import Basemap
#from scipy.spatial import cKDTree
from scipy.interpolate import NearestNDInterpolator
```

```python
nc = Dataset('air.mon.ltm.nc')
hola = nc.variables['air'][0,:]
lons = nc.variables['lon'][:]
lats = nc.variables['lat'][:]
```

```python
# I create a 2D function
def my_func(x, y):
    return exp(-((x-0.5)**2)/0.1) * cos(15.0*(y**2))
```

```python
# show on 500 points
points = np.random.rand(shape(hola)[0],shape(hola)[1])
values = my_func(points[:,0], points[:,1])
print values
```

```python
#here i define a dense grid in x-y with ranges [0,1], [0,1] to see the complete funcion
grid_x, grid_y = meshgrid(lons, lats)
```

```python
grid_zA = NearestNDInterpolator(points, hola)
new_lons = grid_zA(hola[0])
lon2d, lat2d = meshgrid(new_lons, lats)
b = Basemap(lon_0=180) #This defines the center of the map, 180 since longitudes go from 0 to 360
x, y = b(lon2d, lat2d)
b.contourf(x, y, hola, cmap=get_cmap('Greens'))
b.drawcoastlines()
b.colorbar();
```
